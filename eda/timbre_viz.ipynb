```python
# assume df has dropped Unnamed: 0 column and transformed timbre and pitch
def graph_nth(df, nth_segment, n_timbre):
    genres = ['Rap', 'Rock', 'Punk', 'RnB', 'Folk', 'Electronic', 'Metal', 'Blues', 'Pop', 'Jazz', 'Latin', 'Country', 'Reggae']

    df = df.groupby('track_id').head(nth_segment).drop_duplicates(subset = ['track_id'], keep = 'last')
    out = {}
    for genre in genres:
        out[genre] = df.loc[df['genre'] == genre]['timbre'].apply(lambda x: np.array(x)).mean(axis = 0)
        
#     print(out)
    
    
    outt = {}
    for key in out:
        outt[key] = out[key].T
    
    vals = {}
    for i in range(12):
        vals[f'timbre_{i}'] = {}
        for genre in genres:
            vals[f'timbre_{i}'][genre] = outt[genre][i]
            
    segment_range = [i for i in range(50 * n_timbre, 50 * (n_timbre + 1))]

    for i in range(n_timbre):
        pd.DataFrame(vals[f'timbre_{i}']).plot()
```

```python
import pandas as pd
import numpy as np
import json
import os

def graph_nth_full_msd(nth_segment, n_timbre):
    
    base = '../../../public/formatted_csv/'
    small_dfs = []
    for dirr in os.listdir(base):
        dfs = []
        for csv in os.listdir(os.path.join(base, dirr)):
            dfs.append(pd.read_csv(os.path.join(base, dirr, csv)))
        small_df = pd.concat(dfs)
        small_df = small_df.groupby('track_id').head(nth_segment).drop_duplicates(subset = ['track_id'], keep = 'last')
        small_dfs.append(small_df)
        print(f'{dirr} done')

    big_df = pd.concat(small_dfs)
    big_df = big_df.drop(columns = ['Unnamed: 0'])
    big_df['timbre'] = big_df['timbre'].transform(json.loads)
    big_df['pitch'] = big_df['pitch'].transform(json.loads)

    
    genres = ['Rap', 'Rock', 'Punk', 'RnB', 'Folk', 'Electronic', 'Metal', 'Blues', 'Pop', 'Jazz', 'Latin', 'Country', 'Reggae']

    df = big_df.groupby('track_id').head(nth_segment).drop_duplicates(subset = ['track_id'], keep = 'last')
    out = {}
    for genre in genres:
        out[genre] = df.loc[df['genre'] == genre]['timbre'].apply(lambda x: np.array(x)).mean(axis = 0)
        
#     print(out)
    
    
    outt = {}
    for key in out:
        outt[key] = out[key].T
    
    vals = {}
    for i in range(12):
        vals[f'timbre_{i}'] = {}
        for genre in genres:
            vals[f'timbre_{i}'][genre] = outt[genre][i]
    return vals
            
    segment_range = [i for i in range(50 * n_timbre, 50 * (n_timbre + 1))]

    for i in range(n_timbre):
        pd.DataFrame(vals[f'timbre_{i}']).plot()
```

```python
vals = graph_nth_full_msd(2, 4)
```

```python
tims = {
    0: 'Loudness',
    1: 'Brightness',
    2: 'Flatness',
    3: 'Attack'
}

for i in range(4):
    a = pd.DataFrame(vals[f'timbre_{i}'])
    a.index = list(range(50, 100))
    ax = a.plot(figsize = (7, 4), title = f'Average for Element {i} of Timbre ({tims[i]}) Per Genre')
    ax.set_xlabel('Segment')
    ax.figure.savefig(f'timbre_{i}.svg')
```

```python
import pandas as pd
import numpy as np
import json
import os


def graph_nth_full_msd_p(nth_segment, n_timbre):
    
    base = '../../../public/formatted_csv/'
    small_dfs = []
    i = 0
    for dirr in os.listdir(base):
        i += 1
        dfs = []
        for csv in os.listdir(os.path.join(base, dirr)):
            dfs.append(pd.read_csv(os.path.join(base, dirr, csv)))
        small_df = pd.concat(dfs)
        small_df = small_df.groupby('track_id').head(nth_segment).drop_duplicates(subset = ['track_id'], keep = 'last')
        small_dfs.append(small_df)
        print(f'{dirr} done')
        
        if (i == 10):
            break
        

    big_df = pd.concat(small_dfs)
    big_df = big_df.drop(columns = ['Unnamed: 0'])
#     big_df['timbre'] = big_df['timbre'].transform(json.loads)
    big_df['pitch'] = big_df['pitch'].transform(json.loads)

    
    genres = ['Rap', 'Rock', 'Punk', 'RnB', 'Folk', 'Electronic', 'Metal', 'Blues', 'Pop', 'Jazz', 'Latin', 'Country', 'Reggae']

    df = big_df.groupby('track_id').head(nth_segment).drop_duplicates(subset = ['track_id'], keep = 'last')
    out = {}
    for genre in genres:
        out[genre] = df.loc[df['genre'] == genre]['pitch'].apply(lambda x: np.array(x)).mean(axis = 0)
        
#     print(out)
    
    
    outt = {}
    for key in out:
        outt[key] = out[key].T
        
    return outt
    
    vals = {}
    for i in range(12):
        vals[f'pitch_{i}'] = {}
        for genre in genres:
            vals[f'pitch_{i}'][genre] = outt[genre][i]
    return vals
            
    segment_range = [i for i in range(50 * n_timbre, 50 * (n_timbre + 1))]

    for i in range(n_timbre):
        pd.DataFrame(vals[f'pitch_{i}']).plot()
```
